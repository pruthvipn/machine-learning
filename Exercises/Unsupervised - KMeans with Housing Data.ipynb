```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
house_data = pd.read_csv('data/house_rental_data.csv', index_col='Unnamed: 0')
```

```python
house_data.head()
```

```python
models = []
cost = []
for i in range(1,20):
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(house_data)
    models.append(km)
    cost.append(-1 * km.score(house_data))
```

```python
cost
```

```python
plt.plot(range(1,20), cost)
```

```python
plt.show()
```

```python
# the elbow in the curve is around 3
km = KMeans(n_clusters=3, random_state=0)
```

```python
km.fit(house_data)
```

```python
classes = km.predict(house_data)
```

```python
house_data['class']= classes
```
