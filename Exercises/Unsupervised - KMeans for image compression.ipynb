```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from skimage.io import imread
```

```python
image = imread('data/pruthvi_pic1.jpg')
```

```python
image.shape
```

```python
image_scaled = image/255.0
```

```python
image.shape
```

```python
image_scaled = image.reshape(526 * 526, 3)
```

```python
km = KMeans(n_clusters=7) # n_clusters is number of colors in image
```

```python
km.fit(image_scaled)
```

```python
pred_colors = km.predict(image_scaled)
```

```python
km.cluster_centers_
```

```python
reduced_colors = km.cluster_centers_[pred_colors]
```

```python
reduced_image = reduced_colors.reshape(526, 526, 3)
```

```python
plt.imshow(image)
plt.show()
```

```python
plt.imshow(reduced_image)
plt.show()
```
